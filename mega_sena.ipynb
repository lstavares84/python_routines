```python
# Dicionário com os jogos de cada pessoa
jogos = {
    "Lucas": [3, 5, 9, 54, 55, 59, 51],
    "Lais": [10, 11, 12, 13, 14, 15],
    "Ricardo": [21, 22, 23, 14, 15, 16],
    "Vitor": [41, 42, 43, 46, 45, 48]
}
# Fim do dicionário com os jogos de cada pessoa



resultado = [3, 5, 9, 54, 55, 59] # Resultado sorteado

# Verifica quantas dezenas cada pessoa acertou
for pessoa, numeros in jogos.items():
    acertos = 0
    for numero in numeros:
        if numero in resultado:
            acertos += 1
    
    if acertos == 6:
        print(f"{pessoa} acertou {acertos} dezenas e fez a MEGA SENA!")
    elif acertos == 5:
        print(f"{pessoa} acertou {acertos} dezenas e fez a quina!")
    elif acertos == 4:
        print(f"{pessoa} acertou {acertos} dezenas e fez a quadra!")
    else:
        print(f"{pessoa} acertou {acertos} dezenas e não teve sorte dessa vez.")
# Fim Verifica quantas dezenas cada pessoa acertou
```

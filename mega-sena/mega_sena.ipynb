```python
import pandas as pd

caminho_arquivo = 'resultado_megasena.xlsx'

df = pd.read_excel(caminho_arquivo, index_col=0)
df.head()
```

```python
# Dicionário com os jogos de cada pessoa
jogos = {
    "jogo1": [10, 25, 21, 23, 38, 44],
    "jogo2": [13, 30, 35, 39, 50, 54],
    "jogo3": [12, 23, 35, 44, 52, 57],
    "jogo4": [10, 13, 23, 25, 42, 56],
    "jogo5": [17, 20, 31, 34, 40, 42]
}

# Convertendo o dicionário em um DataFrame
df_jogos = pd.DataFrame(jogos)

# Verificando a presença das combinações nos resultados da Mega Sena
for jogo, numeros in jogos.items():
    # Verifica se pelo menos uma combinação está presente em alguma linha do DataFrame original
    resultado_encontrado = df[df.iloc[:, 1:].isin(numeros).all(axis=1)]
    
    if not resultado_encontrado.empty:
        print(f"Combinação do {jogo} encontrada no(s) concurso(s):")
        print(resultado_encontrado.index.tolist())
        print("\n")
    else:
        print(f"Combinação do {jogo} não encontrada.\n")
```

```python
# Convertendo o dicionário em um DataFrame
df_jogos = pd.DataFrame(jogos)

# Inicializando um DataFrame vazio para armazenar os resultados
resultados = pd.DataFrame(index=df.index)

# Iterando sobre cada jogo
for jogo, numeros in jogos.items():
    # Verificando a quantidade de acertos para cada concurso
    acertos = df[df.iloc[:, 1:].isin(numeros)].count(axis=1)
    
    # Adicionando os resultados ao DataFrame
    resultados[f"{jogo}_acertos"] = acertos

# Filtrando os resultados para mostrar apenas aqueles com 4, 5 ou 6 acertos
resultados_filtrados = resultados[(resultados >= 4) & (resultados <= 6)].dropna(how="all")

# Exibindo os resultados filtrados
resultados_filtrados.insert(0, 'Data', df['Data'])
resultados_filtrados.dropna
resultados_filtrados
```

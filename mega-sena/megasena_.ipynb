```python
import pandas as pd
from itertools import combinations
import random

# Função para gerar combinações de dezenas obrigatórias
def generate_combinations(dezenas_obrigatorias, num_dezenas_escolhidas):
    return combinations(dezenas_obrigatorias, num_dezenas_escolhidas)

# Adicionar jogos fixos e combinações à lista de jogos
jogos_fixos = {
    "jogo1": [10, 15, 21, 23, 38, 44],
    "jogo2": [13, 30, 35, 39, 50, 54],
    "jogo3": [12, 23, 35, 44, 52, 57],
    "jogo4": [10, 13, 23, 25, 42, 56],  
    "jogo5": [5, 8, 10, 13, 14, 22],
    "jogo_teste": [17, 20, 31, 34, 40, 42]
}

# Adicionar combinações geradas dinamicamente à lista de jogos
num_dezenas_escolhidas = 6
combinacoes_geradas = list(generate_combinations(dezenas_obrigatorias, num_dezenas_escolhidas))

for i, combinacao in enumerate(combinacoes_geradas, start=1):
    jogos_fixos[f"combinacao_{i}"] = list(combinacao)

# Converter a lista de jogos em um DataFrame
df_jogos = pd.DataFrame(jogos_fixos)

# Lógica para verificar as combinações nos resultados da Mega Sena
df = pd.read_excel('resultado_megasena.xlsx', index_col=0)

# Inicializando um DataFrame vazio para armazenar os resultados
resultados = pd.DataFrame(index=df.index)

# Iterando sobre cada jogo gerado dinamicamente e cada combinação
for jogo, numeros in jogos_fixos.items():
    # Verificando a quantidade de acertos para cada concurso
    acertos = df[df.iloc[:, 1:].isin(numeros)].count(axis=1)
    
    # Adicionando os resultados ao DataFrame
    resultados[f"{jogo}_acertos"] = acertos

# Filtrando os resultados para mostrar apenas aqueles com 4, 5 ou 6 acertos
resultados_filtrados = resultados[(resultados >= 4) & (resultados <= 6)].dropna(how="all")

# Exibindo os resultados filtrados
resultados_filtrados.insert(0, 'Data', df['Data'])
resultados_filtrados = resultados_filtrados.dropna()
print(resultados_filtrados)
```

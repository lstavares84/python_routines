```python
def dijkstra(start, end, distancias, tabela_preco_diesel):
    # Verifique se o n√≥ de in√≠cio e o n√≥ de destino est√£o presentes na tabela de dist√¢ncias
    if start not in distancias or end not in distancias:
        print("N√≥ de in√≠cio ou destino n√£o encontrado na tabela de dist√¢ncias.")
        return None, None

    # Vari√°veis para armazenar a dist√¢ncia total (f) e a dist√¢ncia acumulada (g) entre as cidades definidas como IN√çCIO e FIM
    g = {node: float('inf') for node in distancias}
    f = {node: float('inf') for node in distancias}

    parents = {node: None for node in distancias}  # n√≥ superior ou n√≥ anterior

    # PREDEFINI√á√ÉO de vari√°veis antes de come√ßar a calcular
    g[start] = 0  # Dist√¢ncia total (f) e dist√¢ncia acumulada (g) a partir do ponto de IN√çCIO
    f[start] = 0  # Dist√¢ncia total (f) e dist√¢ncia acumulada (g) a partir do ponto de IN√çCIO
    closed = set()  # Vari√°vel para armazenar os caminhos poss√≠veis j√° testados
    open_set = {start}  # Vari√°vel para armazenar os caminhos poss√≠veis A SEREM testados

    # Vari√°veis para armazenar os caminhos com a menor dist√¢ncia e menor custo
    menor_distancia_caminho = None
    menor_custo_caminho = None

    # Vari√°veis para armazenar a menor dist√¢ncia total e menor custo total
    menor_distancia_total = float('inf')
    menor_custo_total = float('inf')

    ### LOOP PARA CALCULAR A MENOR DIST√ÇNCIA
    while open_set:  # Enquanto houver caminhos n√£o testados...
        # Verifica o caminho atual com a menor dist√¢ncia entre a cidade atual e a cidade vizinha e continue procurando
        current = min(open_set, key=lambda node: f[node])

        # Verifica se o n√≥ atual n√£o √© None
        if current is None:
            break

        # SE a cidade atual (n√≥) for o destino (FIM), calcule do fim at√© o in√≠cio o caminho completo usando os n√≥s pais
        if current == end:
            path = []
            while current:
                path.append(current)
                current = parents[current]
            path.reverse()

            # Calcula a dist√¢ncia total percorrida
            total_distance = g[end]

            # Calcula o custo total percorrido
            total_cost = 0

            # Calcula e imprime as dist√¢ncias entre os n√≥s ao longo do caminho, considerando os valores da tabela de pre√ßos do diesel
            diesel_costs = []
            for i in range(len(path) - 1):
                node1 = path[i]
                node2 = path[i + 1]
                distance = distancias[node1][node2]
                diesel_cost = distance * tabela_preco_diesel[node1]
                total_cost += diesel_cost  # Adiciona o custo ao custo total
                diesel_costs.append((f'Dist√¢ncia de {node1} para {node2}: {distance}', f'Custo de diesel: {diesel_cost}'))

            # Verifica se este caminho tem a menor dist√¢ncia total
            if total_distance < menor_distancia_total:
                menor_distancia_total = total_distance
                menor_distancia_caminho = path.copy()

            # Verifica se este caminho tem o menor custo total
            if total_cost < menor_custo_total:
                menor_custo_total = total_cost
                menor_custo_caminho = path.copy()

            # Imprime a solu√ß√£o na tela para o usu√°rio
            # print('\n\n############## C√ÅLCULO ##############')
            # print(f'O caminho mais curto de {start} para {end} √©:')
            # print(f'Caminho: {path}')
            # print(f'Dist√¢ncia total percorrida: {total_distance}')
            # print(f'Custo total percorrido: {total_cost}\n')
            # print('Dist√¢ncias e custos de diesel entre os n√≥s ao longo do caminho:')
            # for distance, cost in diesel_costs:
            #     print(distance)
            #     print(cost)

            break  # Saia do loop assim que encontrar o destino

        # FIM DE # SE a cidade atual (n√≥) for o destino (FIM), calcule do fim at√© o in√≠cio o caminho completo usando os n√≥s pais

        # Para mover o n√≥ testado de open_set (n√£o testado) para closed (caminhos testados)
        open_set.remove(current)
        closed.add(current)

        # Identificar todas as cidades vizinhas da cidade/n√≥ atual
        for neighbor in distancias[current]:
            if distancias[current][neighbor] is None:
                continue  # Ignora valores None

            tentative_g = g[current] + distancias[current][neighbor]  # Calcular a dist√¢ncia acumulada do n√≥/cidade atual para o n√≥/cidade vizinho selecionado

            # SE a dist√¢ncia acumulada at√© o n√≥/cidade selecionado for menor (mas n√£o igual) √† dist√¢ncia acumulada atual, por favor, atualize "g[vizinho]"
            # e "f[vizinho]".
            if tentative_g < g[neighbor]:
                parents[neighbor] = current  # Define o n√≥/cidade atual como pai do vizinho atual
                g[neighbor] = tentative_g  # atualiza a dist√¢ncia acumulada
                f[neighbor] = g[neighbor]  # estima a dist√¢ncia total restante para chegar √† cidade/n√≥ final

                # Teste se o vizinho atual n√£o foi testado antes. Mova o vizinho atual testado de n√£o testado
                if neighbor not in open_set:
                    open_set.add(neighbor)
                # FIM DE Teste se o vizinho atual n√£o foi testado antes. Mova o vizinho atual testado de n√£o testado

    # Imprime o caminho com a menor dist√¢ncia total e o menor custo total
    print('\n\n############## SOLU√á√ÉO ##############')
    print(f'O caminho mais curto de {start} para {end} com a menor dist√¢ncia √©:')
    print(f'Caminho: {menor_distancia_caminho}')
    print(f'Dist√¢ncia total percorrida: {menor_distancia_total}\n')

    print(f'O caminho mais curto de {start} para {end} com o menor custo √©:')
    print(f'Caminho: {menor_custo_caminho}')
    print(f'Custo total percorrido: {menor_custo_total}\n')

    return menor_distancia_caminho, menor_custo_caminho

# Para cada combina√ß√£o de start e end
for start in capitais:
    for end in capitais:
        if start != end:  # Evitar combina√ß√µes iguais de start e end
            print(f"Calculando caminho de {start} para {end}...\n")
            shortest_distance_path, shortest_cost_path = dijkstra(start, end, distancias, tabela_preco_diesel)
            print("-----------------------------------------------------------\n")
            
            # Verifique se os caminhos s√£o diferentes
            if shortest_distance_path != shortest_cost_path:
                print(f"Os caminhos de {start} para {end} t√™m resultados diferentes!")
                break
```

```python
import pandas as pd

# Carrega o arquivo lotofacil.xlsx
df = pd.read_excel('lotofacil.xlsx')

df
```

```python
import random

# Lista de bolas para procurar
#lista = [1, 3, 4, 9, 10, 14, 15, 16, 17, 18, 20, 21, 22, 23, 25] #nÃºmeros aleatÃ³rios da "sorte"
# lista = [22, 16, 21, 24, 6, 10, 18, 5, 20, 4, 12, 13, 23, 3, 25] #nÃºmeros jÃ¡ sorteados previamente para teste
lista = random.sample(range(1, 26), 15) #gerador aleatÃ³rio
lista = sorted(lista)  # por em ordem crescente

print(lista)
print(f'Total de dezenas na lista: {len(lista)}')

# Converte a lista para um conjunto para facilitar a verificaÃ§Ã£o de pertencimento
conjunto_bolas = set(lista)

# Itera sobre as linhas do DataFrame
for index, row in df.iterrows():
    # ObtÃ©m as bolas do concurso atual
    bolas_concurso = set(row['bola {}'.format(i)] for i in range(1, 16))

    # Verifica a quantidade de valores da lista presentes no sorteio atual
    quantidade_sorteados = len(conjunto_bolas.intersection(bolas_concurso))

    # Verifica se a quantidade de valores sorteados estÃ¡ entre 11 e 15
    if 13 <= quantidade_sorteados <= 15:
        # Imprime o nÃºmero do concurso e a quantidade de dezenas sorteadas
        print("Concurso: {}, Quantidade de dezenas sorteadas: {}".format(row['Concurso'], quantidade_sorteados))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


tamanho = 50 #[50, 500, 1000, 5000]
#random_numbers = np.array(np.random.choice(tamanho, size=tamanho, replace=False))
random_numbers = list(range(tamanho-1, -1, -1))
print("Lista aleatório:", random_numbers)
```

```python
#SELECTION SORT

def selection_sort(arr):
    numero_trocas = 0
    contador_de_loops = 0
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
                numero_trocas = numero_trocas + 1
        arr[i], arr[min_index] = arr[min_index], arr[i]
        print(f"Passo {i}: {arr}")
        contador_de_loops = contador_de_loops + 1
    print("Número de trocas:", numero_trocas)  
    print("Número de Loops:", contador_de_loops)  
    return arr

#random_numbers = np.random.choice(tamanho, size=tamanho, replace=False)
random_numbers1 = random_numbers.copy()
print("Lista aleatório:", random_numbers1)

ordered_numbers = selection_sort(random_numbers1)
print("Lista ordenada:", ordered_numbers)
```

```python
#INSERTION SORT

def insertion_sort(arr):
    numero_trocas = 0
    contador_de_loops = 0
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
            numero_trocas = numero_trocas + 1
        arr[j + 1] = key
        print(f"Passo {i}: {arr}")  # Imprimir a lista a cada passo do loop i
        contador_de_loops = contador_de_loops + 1
    print("Número de trocas:", numero_trocas)  
    print("Número de Loops:", contador_de_loops)  
    return arr

random_numbers2 = random_numbers.copy()
#random_numbers = np.random.choice(tamanho, size=tamanho, replace=False)
print("Lista aleatório:", random_numbers2)

ordered_numbers = insertion_sort(random_numbers2)
print("Lista ordenada:", ordered_numbers)
```

```python
# QUICKSORT
def quicksort(arr, left, right):
    global contador_de_loops  # Declarando como global para ser modificada
    if left < right:
        pivot_index = partition(arr, left, right)
        quicksort(arr, left, pivot_index - 1)
        quicksort(arr, pivot_index + 1, right)
        contador_de_loops += 1  # Incrementa o contador de loops

def partition(arr, left, right):
    global numero_trocas  # Declarando como global para ser modificada
    pivot = arr[right]
    i = left - 1
    global contador_de_passos
    for j in range(left, right):
        contador_de_passos += 1
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
            numero_trocas += 1  # Incrementa o contador de trocas
        print(f"Passo {contador_de_passos}: {arr}, Pivot: {pivot}, Left: {left}, Right: {right}")

    arr[i + 1], arr[right] = arr[right], arr[i + 1]
    return i + 1

random_numbers3 = random_numbers.copy()
numero_trocas = 0
contador_de_loops = 0
contador_de_passos = 0
print("Lista aleatória:", random_numbers3)

quicksort(random_numbers3, 0, len(random_numbers3) - 1)
print("Lista ordenada:", random_numbers3)
print("Número de trocas:", numero_trocas)
print("Contador de loops:", contador_de_loops)
```

```python
def shellsort(arr):
    global numero_trocas, contador_de_passos, contador_de_loops
    n = len(arr)
    gap = n // 5
    while gap > 0:
        print("\n\n#### SUBARRAYS ####")
        for i in range(gap):
            sublist = arr[i::gap]
            print(f"Subarray {i+1}")
            print(sublist)
        print("\n\n#### ORDENANDO OS ARRAYS ####")
        for i in range(gap, n):
            contador_de_passos += 1
            print(f'Passo {contador_de_passos}')

            temp = arr[i]
            j = i
            while j >= gap and arr[j - gap] > temp:
                arr[j] = arr[j - gap]
                j -= gap
                numero_trocas += 1
            arr[j] = temp
            print(arr)
            contador_de_loops += 1
        gap //= 2
    return arr

# Criando um array reverso de 0 a 999
random_numbers4 = random_numbers.copy()
numero_trocas = 0
contador_de_loops = 0
contador_de_passos = 0
print("Lista aleatória:", random_numbers4)

ordered_numbers = shellsort(random_numbers4)
print("Lista ordenada:", ordered_numbers)

print("Número de trocas:", numero_trocas)
print("Número de Loops:", contador_de_loops)
print("Número de Passos:", contador_de_passos)
```
